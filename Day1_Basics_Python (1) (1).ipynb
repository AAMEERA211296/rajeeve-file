# ***MINI CHALLENGE***



```python
students = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 91,
    "David": 68,
    "Eva": 79
}
print(students)

```

```python
average = sum(students.values()) / len(students)
print("Average Marks:", average)
```

```python
highest_student = max(students, key=students.get)
lowest_student = min(students, key=students.get)
```

```python
print("\nHighest Scorer:")
print(highest_student, "-", students[highest_student])
```

```python
print("\nLowest Scorer:")
print(lowest_student, "-", students[lowest_student])
```

```python
print("\nStudents Scoring Above Average:")
for name, marks in students.items():
    if marks > average:
        print(name, "-", marks)
```

```python
search_name = input("\nEnter student name to search: ")

if search_name in students:
    print(search_name, "scored", students[search_name], "marks.")
else:
    print("Student not found.")
```

```python
update_name = input("\nEnter student name to update marks: ")

if update_name in students:
    new_marks = int(input("Enter new marks: "))
    students[update_name] = new_marks

    # Recalculate average
    new_average = sum(students.values()) / len(students)

    print("\nMarks Updated Successfully!")
    print("Updated Student Records:")
    for name, marks in students.items():
        print(name, ":", marks)

    print("Revised Average Marks:", new_average)
else:
    print("Student not found.")
```
